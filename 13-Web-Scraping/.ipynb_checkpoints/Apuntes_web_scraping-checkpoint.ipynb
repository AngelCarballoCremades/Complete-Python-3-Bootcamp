```python
two_stars_books = []
pages = list(range(1,51))
base_url = 'http://books.toscrape.com/catalogue/category/books_1/page-{}.html'

for page in pages:
    res = requests.get(base_url.format(page))
    soup = bs4.BeautifulSoup(res.text, 'lxml')
    products = soup.select('.product_pod')
    
    for book in products:
        if book.select('.star-rating.Two') != []:
            title = book.select('a')[1]['title']
            two_stars_books.append(title)
        else:
            pass

for book in two_stars_books:
    print(book)
```
